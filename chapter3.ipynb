```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next_node=None
        self.prev_node=None
        
class MyStack2:
    def __init__(self):
        self.peek=None
    
    def push(self,value):
        if self.peek is None:
            self.peek=Node(value)
        else:
            new_node=Node(value)
            new_node.next_node=self.peek
            self.peek=new_node
            
    def pop(self):
        if self.peek is None:
            return None
        else:
            value=self.peek.value
            self.peek=self.peek.next_node
            return value
            
    def has_next(self):
        if self.peek is None:
            return False
        else:
            return True

# not in use
class MyStack:
    def __init__(self):
        self.data=[]
        
    def push(self,value):
        self.data.append(value)
        
    def pop(self,value):
        if len(self.value)>0:
            return None
    
        popped_value=self.data[-1]
        self.value=self.value[:-1]
    
    def head(self):
        return self.data[-1]
    
    def has_next():
        if len(self.data)>0:
            return True
        else:
            return False
        
#test
stack=MyStack2()
stack.push(0)
stack.push(1)
stack.push(9)
while stack.has_next():
    print stack.pop()
```

```python
#3.4
class HanoiTower:
    def __init__(self):
        self.towers=[MyStack2() for i in range(3)]
        self.height=0
        
    def initialize_problem(self,height):
        for i in range(height):
            self.towers[0].push(i)
            self.height=height

    def solve(self):
        self.solve_recursive(self.height,self.towers[0],self.towers[1])
    
    def solve_recursive(self,height,from_,to_):
        if height==1:
            to_.push(from_.pop())
        else:
            neutral_tower=self.get_neutral_tower(from_,to_)
            self.solve_recursive(height-1,from_,neutral_tower)
            to_.push(from_.pop())
            self.solve_recursive(height-1,neutral_tower,to_)
    
    def get_neutral_tower(self,from_,to_):
        for tower in self.towers:
            if tower is not from_ and tower is not to_:
                return tower

    def dump_towers(self):
        while True:
            all_empty=True
            content=""
            for tower in self.towers:
                all_empty&= not tower.has_next()
                content+=str(tower.pop())+" "
            
            if all_empty:
                return
            print content
#test
print "#case1"
ht=HanoiTower()
ht.initialize_problem(10)
ht.dump_towers()

print "#case2"
ht=HanoiTower()
ht.initialize_problem(10)
ht.solve()
ht.dump_towers()
            
```
